### Script for plotting yearly plots 

```python
def YearlyPlots(T,ttl='',yl='',xl='',y=None,x=None,size=(10,7), c=None):
    yearday=[i for i in range(1,366)]
    fig=figure(1,figsize=size,dpi=300)
    if shape(T)[0] != 365:
        raise ValueError("First dimension of T should be 365. Shape(T)="+str(shape(T)))
    if c is not None:
        plot_date(yearday,T, '-',color=c);
    else:
        plot_date(yearday,T, '-', );
    # rotate and align the tick labels so they look better
    #fig.autofmt_xdate()
    ylabel(yl)
    xlabel(xl)
    if y is not None:
        ylim(y)
    if x is not None:
        xlim(x)
    grid()
    title(ttl)
```

### Plot the following 3 plots for each measurement:

1. A histogram from the sample values (from SortedVals) restricted between low100 and high100 (By which we mean that any value larger or equal to low100 and smaller or equal to high100 is included).
2. Plot of mean and mean $\pm$ std
3. Number of measurements recorded each day

```python
figure(figsize=(15,30))
offset=1
for meas in STAT.keys():
    subplot(6,3,offset)
    offset += 1
    selection= STAT[meas]['SortedVals'][(STAT[meas]['SortedVals'] <= STAT[meas]['high100']) \
                                        & (STAT[meas]['SortedVals'] >= STAT[meas]['low100'])]
    plt.title(meas + ' restricted histogram')
    plt.hist(selection, 100, facecolor = (0,0,1), alpha=0.75)
    
    ## Your code for mean and mean +- std
    subplot(6,3,offset)
    offset += 1
    std = sqrt(STAT[meas]["Var"])
    Mean = STAT[meas]['Mean']
    YearlyPlots(Mean, ttl = '%s Mean'%meas,yl='',xl='', y = None, x = None,size=(10,7), c = (1, 0 , 0))
    YearlyPlots(Mean + std, ttl = '%s Mean +- STD'%meas, yl = '',xl='',y = None,x = None, size = (10,7), c = (0, 0 , 1))
    YearlyPlots(Mean - std, ttl = '%s Mean +- STD'%meas, yl = '',xl='',y = None,x = None, size = (10,7), c = (0, 0, 1))
    
    ## Your code for number of measurements
    subplot(6,3,offset)
    offset += 1
    YearlyPlots(STAT[meas]['NE'], ttl= meas + " counts", c= (0,0,0))
```
